```python
# Desenvolvido em python (jupyter)

#
# Neste cdigo  plotado o mapa mundi. Em especial foram plotadas bolinhas representando os semforos de Recife. 
# A cor de cada bolinha est relacionada com o valor do sinal referente a cada semforo. No caso em questo
# tem-se que o sinal varia de 0 at 100, a relao entre o valor do sinal e a cor est ilustrada na barra de cores 
# acima do grfico. Os valores atribudos a cada semforo so apenas para mostrar que o cdigo funciona.
#


# Para que o cdigo funcione  precio instalar as bibliotecas abaixo. 
# "folium" e "branca.colormap" so responsveis pela plotagem do mapa 


###
import folium
import csv
import numpy as np
import branca.colormap as cm
import random
from collections import defaultdict
            

###
# Abrindo o arquivo txt com as coordenadas e armazenando os valores numa matriz
columns = defaultdict(list) 

with open('lombada.csv') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        for (i,v) in enumerate(row):
            columns[i].append(v)
            

l = len(columns[0])



# Coordenadas de Recife
RECIFE_COORD = [-8.05389, -34.88111]


# Build map 
#  possvel alterar o estilo visual do mapa atravs do parmetro 'tiles'
map_nyc = folium.Map(location=RECIFE_COORD, zoom_start=12, tiles='cartodbpositron', width=640, height=480)

# Pegando as coordenadas dos semforos
coords = np.zeros((l,2))

for p in range(0,l):
    coords[p] = [columns[2][p],columns[3][p]]
    
coords = list(coords)

# Aps seguir um passo a passo (conforme links.txt) obtm-se as coordenadas das bordas do mapa do Recife
# Na linha 1 do recife.txt esto as coordenadas, sendo preciso armazenar esta linha e em seguida process-la 

# O txt  aberto para leitura
f = open('recife.txt', "r")

# usa-se readlines() para ler o arquivo linha por linha
lines = f.readlines()

# fecha-se o arquivo aps todas as linhas terem sido lidas
f.close()

# Pegando a linha 1
a = lines[1].split(',')

# Criando array de numpy
coordinates = np.zeros((1785,2))

# O range abaixo  determinado, conforme o arquivo, para pegar apenas coordenadas geogrficas da linha 1
# Os ndices nas latitudes so 5, 7, 9 .. e na longitude so 6, 8, 10 .. Por isso usou-se uma PA de grau 1
for i in range(2,1787):
    coordinates[i-2] = [float(a[2*i+1]),float(a[4+2*(i-1)])]    

################################################################
# So atribudos valores aleatrios de 0 a 100 a cada semforo, apenas para ilustrar o funcionamento do cdigo
s = np.zeros((len(coords),1))

for i in range(0,len(coords)):
    s[i] = random.randint(0, 100)
################################################################

# Cria-se um objeto (aux) para obter as cores da barra (RdBu_04) e as cores so armazenadas em uma lista (color = [])

color = []

a = cm.linear.RdBu_04.scale(0, 100)

for i in range(0,len(coords)):
    color.append(a.rgb_hex_str(2*i+20))


# Plotando os semforos no mapa
for i in range(0,len(coords)):
    folium.CircleMarker(coords[i], radius=2, fill_opacity=1, color=color[i], fill_color=color[i]).add_to(map_nyc) 

# Plotando a borda do mapa
folium.PolyLine(locations=coordinates,weight=1,color = 'red').add_to(map_nyc)

# Plota-se agora a barra de cores
colormap = cm.linear.RdBu_04.scale(0, 100)
colormap.caption = 'A colormap caption'
map_nyc.add_child(colormap)


# Display map in Jupyter
map_nyc
```
